```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread("img\LED2_2_2021-11-15-174526-0000.jpg",cv2.IMREAD_GRAYSCALE)
print(img.shape)
```

```python
avg_rows = np.mean(img,axis=1)
print(avg_rows.shape)
```

```python
print(avg_rows)
```

```python
plt.plot(avg_rows)
```

```python
specs = np.fft.rfft(avg_rows)
specs = np.abs(specs)
specs = specs[1:10]
plt.plot(specs)
plt.xlabel("Freq(Hz)")
plt.ylabel("Ampli(-)")
plt.show()
print(sorted(specs,reverse=True)[:5])
```
