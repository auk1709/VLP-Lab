```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
video = cv2.VideoCapture('video/square30-01.mp4')
print(video.isOpened())
```

```python
ret, frame = video.read()
print(ret)
print(frame.shape)
```

```python
rows = []
while ret == True:
    frame = frame[:,:,2]
    frame = np.mean(frame,axis=0)
    rows.append(frame)
    ret, frame = video.read()
rows = np.array(rows)
print(rows.shape)
```

```python
print(rows[0])
```

```python
coefs = [np.corrcoef(rows[0], row)[0,1] for row in rows]
coefs = np.array(coefs)
print(coefs.shape)
```

```python
plt.plot(coefs)
plt.show
```

```python
print(np.argsort(coefs)[::-1][:20])
```
